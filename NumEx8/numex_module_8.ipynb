Affine transforms
-----------------

We have seen in the lecture that an affine transform simply defines a change of the coordinate system. In the most general terms, given a $2 \times 2$ matrix $\mathbf{A}$ and a $2 \times 1$ vector $\mathbf{d}$, an affine transform is expressed by

$\left( \begin{array}{c} y_1 \\ y_2 \end{array} \right) =
\mathbf{A} \left( \begin{array}{c} x_1 \\ x_2 \end{array}
\right) - \mathbf{d}.$ 

The difficulty comes from the fact that we work in a discrete space. Indeed, a pixel of an image is indexed by two integers corresponding to index of the row and the column where this pixel is located. Then, in the formula above, $x_1$ and $x_2$ are integers. In general, since we work with a general matrix $\mathbf{A}$ and vector $\mathbf{d}$, $y_1$ and $y_2$ will not necessarily be integers anymore. In this example, we look at a possible implementation of affine transforms to see how this difficulty is circumvented. 

The solution is already outlined in the lecture slides. For each pixel of the new image, indexed by $y_1$ and $y_2$, we inverse the transform 


$\left( \begin{array}{c} x_1 \\ x_2 \end{array} \right) =
\mathbf{A}^{-1} \left( \left( \begin{array}{c} x_1 \\ x_2
\end{array} \right) + \mathbf{d} \right).$

Then, we find the nearest integers strictly smaller than $x_1$ and $x_2$, referred to as $\eta_1$ and $\eta_2$ respectively and we denote by $\theta_i$ the difference between $x_i$ and $\eta_i$, $i=1,2$. The intensity of the pixel is then obtained using an interpolation formula. For example, with a bilinear interpolation, the intensity of the pixel of the new image $I'$ given the original image $I$ is expressed by:

$I'(y_1, y_2) = (1 - \theta_1)(1 - \theta_2)I(\eta_1, \eta_2) +
\theta_1(1 - \theta_2)I(\eta_1 + 1, \eta_2) + \theta_2(1 -
\theta_1)I(\eta_1, \eta_2 + 1) + \theta_1 \theta_2 I(\eta_1 + 1,
\eta_2 + 1)$ <br>

Finally, it may happen that $\eta_i$ or $\eta_i+1$ fall outside the range of the original image. We then use zero-padding, i.e., we assume that the intensity of the pixels outside the original image is set to 0. 

The following code implements a particular affine transform called shearing.

```python
import numpy as np

# Parameters
S = 0.3

# Construct and display the original image
I_orig = np.zeros((64,64))
I_orig[24,16:48] = 1
I_orig[39, 16:48] = 1;
I_orig[25:56, 16] = 1;
I_orig[8:40, 47] = 1;

# Computing A and d matrixes
A = np.array([[1, S],[0,1]])
d = np.array([[0],[0]])

from numpy import linalg as l

# Initialization
nb_row = I_orig.shape[0]
nb_col = I_orig.shape[1]
I = np.zeros((nb_row,nb_col))

for y1 in range(0,nb_row):
    for y2 in range(0,nb_col):
        x = np.dot(l.inv(A),np.array([[y1],[y2]]) + d)
        eta = np.floor(x)
        theta = x - eta
        
        if int(eta[0]) >= 1 & int(eta[0]) <= nb_row:
            if int(eta[1]) >= 1 & int(eta[1]) <= nb_row:
                I[y1, y2] = I[y1, y2] + (1 - theta[0])*(1 - theta[1])*I_orig[int(eta[0]) - 1, int(eta[1]) - 1]
                
            if int(eta[1]) >= 1 & int(eta[1]) <= (nb_col - 1):
                I[y1, y2] = I[y1, y2] + (1 - theta[0])*theta[1]*I_orig[int(eta[0]) - 1, int(eta[1])];
        
        
        if int(eta[0]) >= 0 & int(eta[0]) <= (nb_row - 1):
            if int(eta[1]) >= 1 & int(eta[1]) <= (nb_col):
                I[y1, y2] = I[y1, y2] + theta[0]*(1 - theta[1])*I_orig[int(eta[0]), int(eta[1]) - 1]
                
            if int(eta[1]) >= 0 & int(eta[1]) <= (nb_col - 1):
                I[y1, y2] = I[y1, y2] + theta[0]*theta[1]*I_orig[int(eta[0]), int(eta[1])]

```

The original image looks like:

```python
%pylab inline
figsize(12,8)
# Showing the image in a gray colormap
import matplotlib.pylab as plt
imshow(I_orig,cmap = plt.get_cmap('gray'))
show()
```

whereas the transformed image is:

```python
# Showing the image in a gray colormap
import matplotlib.pylab as plt
imshow(I,cmap = plt.get_cmap('gray'))
show()
```

Unsharp masking
-------------------

Unsharp masking is an edge enhancement technique commonly available on commercial image manipulation softwares like Photoshop. The principle underlying unsharp masking is quite simple. In its most basic form, it consists in subtracting from the original image a blurred version of the image. The latter is obtained by applying a 2D Gaussian blur filter, whose impulse response is given by:

$h(n_1, n_2) = \frac{1}{2\pi \sigma^2} \exp\left(-\frac{n_1^2 +
n_2^2}{2\sigma^2} \right)$

for $|n_1, n_2| \leq N$. This is thus a simple linear filter applied on the image.


Let us look at the effect of this filter on the following image:

![](http://www.math.ust.hk/~masyleung/Teaching/CAS/MATLAB/image/images/cameraman.jpg)

Download the image from here and save it in your Python workspace directory. Then load the image in the workspace using the `pylab.imread(...)` function:


```python
import matplotlib.pylab as plt

I = np.array(plt.imread('cameraman.jpg'), dtype=float64)
imshow(I, cmap = plt.get_cmap('gray'), interpolation='none')
title('original')
show()
```

Then, we define the impulse response of the Gaussian filter, parametrized by $N$ and $\sigma$,

```python
SIGMA = 12.
N = 10

n1, n2 = np.meshgrid(np.arange(-N,N+1), np.arange(-N,N+1))
h = np.exp(-(n1**2+n2**2)/(2*SIGMA**2))/(2*np.pi*SIGMA**2)
```

The image is then filtered and the result subtracted to the original image:

```python
from scipy import signal as sc
from scipy import ndimage as ip

# we use an optimized convolution routine provided by the ndimage package of scipy
# scipy.signal.convolve2d is just too slow.
I_blurred = ip.convolve(I, h)

I_sharpened = I - I_blurred
```

The resulting image looks like:

```python
imshow(I_sharpened, cmap = plt.get_cmap('gray'), interpolation='nearest')
title('sharpened')
show()
```

By subtracting a blurred, i.e., low-pass fltered version of the image, this procedure boosts the high frequency components of the image. The difference is particularly important in the texture of the grass or the edge of the tower in the background. This procedure also creates artifacts, for example there is a sort of halo that surrounds the cameraman, particularly behind his coat. As this procedure is uniformly applied on all parts of the image, we create the impression of an edge enhancement, but the information content in the image is not improved.

DCT transform and JPEG compression
----------------------------------

In the JPEG compression standard, the image is projected onto another space using the discrete cosine transform (DCT). In this space, the information in the image is concentrated on few coefficients. Most of the transform coefficients can then be discarded, as they convey little additional information. There are of course other technical aspects of great practical interest, like smart quantization or transform coefficients coding. But this simple idea is at the heart of the JPEG compression standard. Let us illustrate it through a simple example.

To run this example, we provide you a Python function that computes the orthonormal basis vectors corresponding to the DCT. First, download a downsampled image of the cameraman from ![https://spark public.s3.amazonaws.com/dsp/num_examples/Module%208/images/camera.jpg](here) and save it in your Python workspace directory. Then load it in the workspace and rearrange it as a column vector.

```python
import matplotlib.pylab as plt

I = np.array(plt.imread('camera_down.jpg'), dtype=float64)
nb_row = I.shape[0]
nb_col = I.shape[1]
```

We then compute the vectors forming an orthonormal basis corresponding to the 2D DCT. These vectors arranged as column vectors form the columns of a basis matrix.

```python
def get_dct2_vectors(N,M):
    # Compute normalized DCT2 vectors
    
    # Initialization
    basis = np.zeros((N*M,N*M))
    X,Y = np.meshgrid(np.arange(N),np.arange(M))
    linX = X.reshape(-1,1)
    linY = Y.reshape(-1,1)
    
    basis = np.cos((np.pi/N)*(linX+0.5)*linY.T)*np.cos((np.pi/M)*(linY+0.5)*linX.T)
    basis /= sqrt((basis**2).sum(axis=0))
            
    return basis.T

basis_matrix = get_dct2_vectors(I.shape[0], I.shape[1])
```

We project the original image onto this new basis and discard half of the coefficients.


```python
I_projected = np.dot(basis_matrix,I.reshape(-1,1))
I_projected[2048:I_projected.size,:] = 0
```

We invert the transform and display the resulting image.

```python
from numpy import linalg as l

I_approx = np.dot(basis_matrix.T, I_projected).reshape(I.shape)

plt.subplot(1,2,1)
imshow(I, cmap = plt.get_cmap('gray'), interpolation='none')
title('original')

subplot(1,2,2)
imshow(I_approx, cmap = plt.get_cmap('gray'), interpolation='none')
title('approximation')
show()
```

We notice that we still recover most of the image, despite the very coarse approximation.
